# Data analysis with Python

## Introductory example

```python
"""As an example, we use the fraser-river flow data which does not have any obvious uniform format.
This gives us an excuse to do see how to read in a file, perform some data munging, calculate simple
statistics and visualize the data."""

text = open('data/fraser-river.txt', 'r').read()
print text
```

```python
"""However we notice that the flows begin with '485' and continue to the end of the file,
separated by spaces or tabs. This is enough information to "munge" the flows into a numpy int array."""

idx = text.find('485')
flows = np.array(text[idx:].strip().split(), dtype='int')
```

```python
"""Let's perform some rudimentary analysis."""

n = len(flows)
mu = flows.mean()
sd = flows.std(ddof=1)
min_flow = flows.min()
max_flow = flows.max()
flood_idx = flows > (mu+2*sd)
outlier_flows = flows[flood_idx]
outlier_dates = np.array(dt)[flood_idx]

print """
Number of observations: %d
Mean flow             : %.2f
Standard deviation    : %.2f
Min flow              : %.2f
Max flow              : %.2f
Mean of outliers      : %.2f
""" % (n, mu, sd, min_flow, max_flow, outlier_flows.mean())

# count number of flodds by month
flood_months = {}
for date in outlier_dates:
    month = date.strftime('%b')
    flood_months[month] = flood_months.get(month, 0) + 1

for month in sorted(flood_months):
    print "Flooded %d times in %s" % (flood_months[month], month)
```

```python
"""A useful step in any data analysis is to visualize the data."""

# first calculate some summary statistics
mu = flows.mean()
sd = flows.std()

# now plot the figure
figure(figsize=(24, 3))
plot(flows)
axhline(mu, color='red')
for val in [mu - sd, mu + sd]:
    axhline(val, color='red', linestyle='dashed')
```

```python
"""Let's pretty up the plot."""

import datetime
import matplotlib.dates as dates
from dateutil import rrule

"""We first need to generate the list of dates where records were taken."""
datemin = datetime.date(1913, 3, 1)
datemax = datetime.date(1990, 12, 1)
dt = list(rrule.rrule(rrule.MONTHLY, dtstart=datemin, until=datemax))
print len(flows) - len(dt)

"""Hmm, there are 12 months short - let's assume that the start year is off by one."""
datemin = datetime.date(1912, 3, 1)
dt = list(rrule.rrule(rrule.MONTHLY, dtstart=datemin, until=datemax))

# now plot the figure again, this time with dates
figure(figsize=(24, 3))
plot(dt, flows)
axhline(mu, color='red')
for val in [mu - sd, mu + sd]:
    axhline(val, color='red', linestyle='dashed')
pylab.ylabel('Mean monthly flows (cm)', fontsize=12)
pylab.title('Fraser River at Hope, B.C.', fontsize=16)
```

```python
"""Now zoom in to the first 48 months to check the most likely flood months."""

start = 0
stop = 48

months   = dates.MonthLocator()  # every month
monthsFmt = dates.DateFormatter('%b %Y')

fig = figure(figsize=(24,3))

ax = fig.add_subplot(111)
ax.plot(dt, flows)
ax.set_xlim(dt[start], dt[stop])

axhline(mu, color='red')
for val in [mu - sd, mu + sd]:
    axhline(val, color='red', linestyle='dashed')
    
pylab.ylabel('Mean monthly flows (cm)', fontsize=12)
pylab.title('Fraser River at Hope, B.C.', fontsize=16)

ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(monthsFmt)

fig.autofmt_xdate()
```
